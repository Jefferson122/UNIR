<b>Ejercicio 4</b><br/>
    Represente las siguientes funciones en una única gráfica:
    <ul>
    <li>$x^2 + 2x - 1$</li>
    <li>$-x^3 + 1$</li>
    <li>$x/2 + 1$</li>
    </ul>
</div>

```python
x = np.arange(-10, 10, 0.1)
y1 = x**2 + 2*x - 1
y2 = -x**3 + 1
y3 = x/2 + 1

plt.plot(x, y1, label="$x^2 + 2x - 1$")
plt.plot(x, y2, label="$-x^3 + 1$")
plt.plot(x, y3, label="$x/2 + 1$")
plt.legend()
plt.show()
```

<p><b>Ejercicio 5</b><br/>
    Represente las siguientes medidas comparativas del rendimiento de procesadores ejecutando un programa (extraído de <a href="https://www.phoronix.com/scan.php?page=news_item&px=i7-1165G7-vs-Ryzen-7-PRO-4750U">Phoronix</a>, los valores menores son mejores). Cree una barra por cada programa:</p>
    <table>
        <tr><th>Procesador</th><th>Programa 1</th><th>Programa 2</th></tr>
        <tr><td>Ryzen 7 4700U</td><td>148.04</td><td>189.62</td></tr>
        <tr><td>Ryzen 5 4500U</td><td>200.77</td><td>243.86</td></tr>
        <tr><td>Core i7 1165G7</td><td>209.92</td><td>263.88</td></tr>
        <tr><td>Core i7 8565U</td><td>312.65</td><td>372.89</td></tr>
    </table>
</div>

```python
procesador = [ "Ryzen 7 4700U", "Ryzen 5 4500U", "Core i7 1165G7", "Core i7 8565U"]
programa1 = [148.04, 200.77, 209.92, 312.65]
programa2 = [189.62, 243.86, 263.88, 372.89]

x = np.arange( len(procesador) )

plt.bar(x-0.1, programa1, label="Programa 1", width=0.2)
plt.bar(x+0.1, programa2, label="Programa 2", width=0.2)
plt.xticks( x, procesador )
plt.legend()
plt.show()
```

```python
figura, ejes = plt.subplots(2,2)
ejes[0][0].plot(x,programa1)
ejes[1][0].plot(x,programa2)
plt.show()
```

Un problema que podemos encontrar es que la gráfica resultante sea demasiado pequeña. Para ello, podemos usar el parámetro <code>figsize</code> para ajustar el tamaño pasando una tupla con la anchira y altura en <a href="https://www.metric-conversions.org/es/longitud/pulgadas-a-centimetros.htm">pulgadas</a>:

```python
figura, ejes = plt.subplots(2,2, figsize=(20,10))
ejes[0][0].plot(x, programa1)
plt.show()
```

Esto nos dará más espacio para poner títulos, etc.

<div style="background-color:lightpink; padding:1em">
<p><b>Ejercicio 6</b><br/>
    Represente las funciones $f(x) = x$, $f(x) = x^2$, $f(x) = x^3$, y $f(x) = x^4$ en una figura con cuatro subfiguras.
</div>

```python
x = np.arange(-10, 10, 0.1)
y1 = x
y2 = x**2
y3 = x**3
y4 = x**4

figura, ejes = plt.subplots(2,2)
ejes[0][0].plot(x, y1)
ejes[0][1].plot(x, y2)
ejes[1][0].plot(x, y3)
ejes[1][1].plot(x, y4)
plt.show()
```
