## Representaciones básicas

### Gráficas de líneas

Empezaremos viendo las representaciones más simples que se pueden hacer. Empezamos con la gráfica de líneas. Para ello, necesitaremos dos objetos <code>ndarray</code> de Numpy (o listas simples de Python), uno que nos diga los valores de la variable independiente ($x$) y otro con los valores de la variable dependiente ($y$), que generaremos de la siguiente manera:

```python
x=np.random.rand(500)
y=0.1*np.random.randn(500)
plt.xlim([-5,5])
plt.ylim([-5,5])
plt.scatter(x,y)
plt.show()
```

```python
x = np.arange(-10,11)
y = x**2
```

```python
print(x)
print(y)
```

En <code>x</code> tendremos los números entre -10 y 10, y en <code>y</code> sus cuadrados. La función para dibujar una gráfica de línea en Matplotlib es la función <code>plot()</code>, documentada <a href="https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html">aquí</a>. Como podemos ver en la siguiente línea, es una función que casi no hace falta explicar. Tras esta función, hacemos una llamada a <code>show()</code>, para que se realice la representación. Esta función no siempre es necesaria en Jupyter, pero es conveniente llamarla para que se realicen los ajustes apropiados y para que podamos exportar el código en el futuro.

```python
plt.plot(x,y)
plt.show()
```

Podemos personalizar muchos parámetros de esta gráfica, por ejemplo, el <a href="https://matplotlib.org/stable/gallery/color/named_colors.html">color</a>, anchura, <a href="https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html">estilo</a> y <a href="https://matplotlib.org/stable/api/markers_api.html">marcadores</a>, pasándole a <code>plot</code> unos parámetros opcionales detallados en la <a href="https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html">documentación</a>:

```python
plt.plot(x,y, color="red", linestyle="dotted", linewidth=2, marker="s")
plt.show()
```

### Nubes de puntos

En análisis de datos, una de las gráficas que mejor representan las relaciones entre dos variables son las nubes de puntos. A diferencia de las gráficas de barras y de líneas, en principio no implicamos que la variable del eje vertical tenga alguna relación con la del eje horizontal (de hecho, usamos este tipo de gráfica para verificar eso). Tampoco hace falta que la variable horizontal tenga valores equiespaciados. En este tipo de gráficas tomamos un valor de cada variable (representado como un <code>ndarray</code> o lista) y los agrupamos en puntos bidimensionales, donde la coordenada horizontal es el valor de la primera variable y la vertical el de la segunda. Veamos un ejemplo con dos variables aleatorias:

```python
x = np.random.randn(10)
y = np.random.randn(10)

print(x)
print(y)
```

La primera entrada de <code>x</code> y la primera de <code>y</code> serán el primer punto, las segundas el segundo y así sucesivamente.

En este caso, la gráfica se hace con <code>scatter</code>, y, nuevamente, su sintaxis básica es la misma, manteniendo la coherencia en Matplotlib:

```python
plt.scatter(x,y)
plt.show()
```

Vemos en este caso que tenemos diez puntos distribuidos aleatoriamente, con lo cual no podemos decir mucho acerca de los datos. Si representamos muchos más, veremos que aparece un patrón que podemos identificar a simple vista:

```python
x = np.random.randn(10000)
y = np.random.randn(10000)

plt.scatter(x,y)
plt.show()
```

Vemos que tenemos una distribución gaussiana bidimensional. Veremos más de esto en el modelado estadístico, pero valga como muestra de que las nubes de puntos nos pueden decir algo acerca de las variables. Por ejemplo, si generamos una variable aleatoria condicionada a otra:

```python
x = np.random.randn(10000)
y = x + np.random.randn(10000)

plt.scatter(x,y)
plt.show()
```

En este caso, hay una cierta relación lineal entre <code>x</code> e <code>y</code>. En análisis de datos, es importante previsualizar este tipo de relaciones, ya que guiarán nuestro trabajo futuro; no podemos confiar ciegamente en los algoritmos de procesado de datos.

## Representación multivariable

Hasta ahora hemos visto representaciones simples de una variable dependiente para las gráficas de líneas y barras. No obstante, podemos mezclar datos de dos variables diferentes que tengan la misma variable independiente. Matplotlib hace esto muy intuitivo al permitir múltiples llamadas a <code>plot()</code>, <code>bar()</code>, etc. Supongamos que tenemos la variable independiente <code>x</code> y las variables dependientes <code>y1</code> e <code>y2</code>:

```python
x = np.arange(-10, 10, 0.1)
y1 = x + 1
y2 = -x**2 + x + 1
```

Si queremos representar <code>y1</code> e <code>y2</code>, no tenemos más que llamar dos veces a <code>plot()</code>:

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

Podemos ver que no sólo Matplotlib sabe representar ambas variables en la misma gráfica, sino que automáticamente elige un color que las distinga. No obstante, puede resultar un poco confuso saber qué representa cada línea sin una leyenda. Podemos añadirla con la función <code>legend()</code>; aunque antes tenemos que decirle a Matplotlib qué queremos poner en esa leyenda, es decir, cómo se etiqueta cada variable. Para ello, pasamos el argumento opcional <code>label</code> a <code>plot()</code>:

```python
plt.plot(x, y1, label="Variable y1")
plt.plot(x, y2, label="Variable y2")
plt.legend()
plt.show()
```

La leyenda se suele colocar en el sitio apropiado de forma automático, pero se puede ajustar su posición manualmente. En el manual de Matplotlib hay un <a href="https://matplotlib.org/stable/tutorials/intermediate/legend_guide.html">apartado dedicado a las leyendas</a>. Otra característica interesante es que podemos usar algunas expresiones de <a href="https://www.latex-project.org/">$\LaTeX$</a> para formatear las leyendas:

```python
plt.plot(x, y1, label="$y_1=x+1$")
plt.plot(x, y2, label="$y_2=x^2+x+1$")
plt.legend()
plt.show()
```

Las gráficas de barras también permiten representar varias variables en una única gráfica. Veamoslo con un ejemplo en el que tenemos datos de ventas de dos tipos de productos de una empresa en función del trimestre:

```python
periodo = ["Q1", "Q2", "Q3", "Q4"]
unidades_producto_1 = [125, 342, 543, 99]
unidades_producto_2 = [348, 288, 198, 245]
```

Veamos qué pasa si intentamos hacer lo mismo que con <code>plot()</code>. Incluimos directamente los parámetros <code>label</code>:

```python
plt.bar(periodo, unidades_producto_1, label="Producto 1")
plt.bar(periodo, unidades_producto_2, label="Producto 2")
plt.legend()
plt.show()
```

En este caso, las barras se taparán unas a otras. Para que no lo hagan, tendremos que representar cada barra en una coordenada $x$ distinta. Esto no se puede hacer directamente; primero tendremos que convertir el eje $x$ actual en uno numérico:

```python
x = np.arange( len(periodo) )
```

A continuación, generaremos dos sub-ejes, uno para cada una de las variables dependientes, de modo que las barras de una estén desviadas a la izquierda, y los otros a la derecha:

```python
x_1 = x - 0.1
x_2 = x + 0.1
```

Veamos cómo queda esto al hacer la representación:

```python
plt.bar(x_1, unidades_producto_1, label="Producto 1")
plt.bar(x_2, unidades_producto_2, label="Producto 2")
plt.legend()
plt.show()
```

Ahora las gráficas se pueden distinguir, pero todavía se superponen. Solventaremos esto ajustando el parámetro de ancho de las barras:

```python
plt.bar(x_1, unidades_producto_1, label="Producto 1", width=0.2)
plt.bar(x_2, unidades_producto_2, label="Producto 2", width=0.2)
plt.legend()
plt.show()
```

Así se ve mucho mejor. Queda un punto por resolver, que es el de las etiquetas del eje X. Veremos esto en el siguiente apartado.

## Personalización del aspecto

Podemos personalizar todos los aspectos de la gráfica. Aquí veremos cómo ajustar el intervalo horizontal y vertical, poner etiquetas a los ejes y a la gráfica, así como personalizar las marcas del eje. 

### Objetos <code>figure</code> y <code>axis</code>

Empecemos por el final, para continuar con el ejemplo de la gráfica de barras. Para poder hacer esto, hay que introducir el concepto de <a href="https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html">figuras</a> y <a href="https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axis.html">ejes</a>, que en Matplotlib nos permiten modificar el comportamiento de la figura en su conjunto (es decir, todo lo que se representa en pantalla) y los ejes (todo lo que atañe al contexto de los datos representados). Por ejemplo, podemos pedirle a Matplotlib que represente dos gráficas en figuras distintas de forma explícita:

```python
plt.bar(x_1, unidades_producto_1, label="Producto 1", width=0.2)
plt.figure()
plt.bar(x_2, unidades_producto_2, label="Producto 2", width=0.2)
plt.legend()
plt.show()
```

En el momento de llamar a la función <code>figure()</code> se genera una nueva figura, y todos los comandos pasan a representar sobre la nueva. Para tener acceso a la figura y al eje, tenemos que llamar a la función <code>subplots()</code>,  de la que recibimos un objeto <code>figure</code> y un <code>axis</code>, que nos permite jugar con estos aspectos. Por ejemplo, podemos explicitar qué etiquetas queremos en un eje llamando a los métodos <code>set_xticks()</code> (que nos dice en qué valores del eje X se representa etiqueta) y <code>set_xticklabels()</code> (que nos permite decir qué etiquetas queremos poner ahí):

```python
figura, eje = plt.subplots()
plt.bar(x_2, unidades_producto_2, label="Producto 2", width=0.2)
eje.set_xticks( x )
eje.set_xticklabels( periodo )
plt.show()
```

Con esto ya tenemos todos los elementos para terminar nuestra representación gráfica:

```python
figura, eje = plt.subplots()
plt.bar(x_1, unidades_producto_1, label="Producto 1", width=0.2)
plt.bar(x_2, unidades_producto_2, label="Producto 2", width=0.2)
eje.set_xticks( x )
eje.set_xticklabels( periodo )
plt.legend()
plt.show()
```

Conseguir la representación exacta de lo que queremos en Matplotlib suele ser un proceso iterativo, en el que siempre está bien tener a mano la <a hreF="https://matplotlib.org/stable/contents.html">documentación</a> de Matplotlib, y, sobre todo, la <a href="https://matplotlib.org/stable/gallery/">galería</a>. La interactividad de Jupyter hace que sea una herramienta muy útil para este tipo de estudios.

Hemos aprovechado esta personalización para ver el concepto de los objetos <code>figure</code> y <code>axis</code>, que nos permitirá personalizar todos los aspectos de la gráfica. No obstante, hay una forma más directa de establecer las etiquetas del eje x, mediante la función <code>xticks()</code>, a la que le pasamos directamente las localizaciones y valores de las etiquetas. Esta función es parte del módulo <code>pyplot</code>, no del objeto <code>axis</code>:

```python
plt.bar(x_1, unidades_producto_1, label="Producto 1", width=0.2)
plt.bar(x_2, unidades_producto_2, label="Producto 2", width=0.2)
plt.xticks( x, periodo )
plt.legend()
plt.show()
```

### Ajuste de zoom en los ejes

La segunda personalización que vamos a ver es cómo podemos ajustar el zoom en el eje horizontal y el vertical. Tomemos una gráfica de líneas de nuevo:

Nota: Matriz de muestras de dos por cuatro de N(3, 6.25) si colocamos 3 + 2.5 * np.random.randn(2, 4)->Matriz de dos pos 4, es una distribución normal

array([[-4.49401501,  4.00950034, -1.81814867,  7.29718677],
       [ 0.39924804,  4.68456316,  4.99394529,  4.84057254]])

```python
x = np.arange(0,10,0.01) #numeros entre 0 y 10 con saltos de 0.01
y = np.random.randn(1000)
y = y + 0.7*np.roll(y, 1) + 0.6*np.roll(y, 2) + 0.5*np.roll(y, 3) 
#np.roll(x,2) lo que hace es correr el vector 2 unidades a la derecha
```

```python
plt.plot(x, y)
plt.show()
```

Esta visualización nos ofrece todos los valores de las variables, si queremos, por ejemplo, hacer zoom entre 0 y 0.5, lo haremos llamando a <code>xaxis</code>:

```python
plt.plot(x, y)
plt.xlim([0, 0.5])
plt.show()
```

También se puede hacer con el objeto <code>axis</code>, pero normalmente usaremos esta forma más simple. Para el eje y, la función que hay que usar es similar:

```python
plt.plot(x, y)
plt.ylim([-10, 10])
plt.show()
```

## Subfiguras

En ocasiones necesitaremos gráficas con varias subfiguras. La función <code>subplots()</code> que vimos antes nos servirá también aquí. En el ejemplo anterior, al haber sólo una subfigura, recibíamos un objeto <code>axis</code> único. Si tenemos varias subfiguras, recibiremos varios <code>axis</code>, y tendremos que usarlos para pintar sobre la gráfica que representan cada uno. Veamos, paso a paso, cómo podemos crear una gráfica con 2 filas y 2 columnas. Lo primero será especificar el formato llamando a <code>subplots(filas, columnas)</code>:

```python
figura, ejes = plt.subplots(2,2)
```

En <code>ejes</code> tendremos una lista de listas de objetos <code>axis</code>. Cada entrada de la lista <code>ejes</code> refleja una fila, y cada fila es una lista de columnas:

<img src="subfig.png" style="width:30em; margin: 0 auto;">

Cada uno de estos objetos <code>axis</code> tiene los mismos métodos de <code>plt</code> para hacer una gráfica en esa subfigura. Por ejemplo, en lugar de llamar a <code>plt.plot()</code>, llamaremos a <code>ejes[0].plot()</code> para representar sobre la primera gráfica:

```python
figura, ejes = plt.subplots(2,2)
ejes[0][0].plot(x, y)
plt.show()
```

Un problema que podemos encontrar es que la gráfica resultante sea demasiado pequeña. Para ello, podemos usar el parámetro <code>figsize</code> para ajustar el tamaño pasando una tupla con la anchira y altura en <a href="https://www.metric-conversions.org/es/longitud/pulgadas-a-centimetros.htm">pulgadas</a>:

```python
figura, ejes = plt.subplots(2,2, figsize=(20,10))
ejes[0][0].plot(x, y)
plt.show()
```

Esto nos dará más espacio para poner títulos, etc.

## Matrices

Con Matplotlib podemos crear un mapa de calor, mediante la función <code>imshow</code>. Esta función pintará cada posición de la matriz según el valor en una escala de colores. Podemos seleccionar una de las muchas escalas de colores que <a href="https://matplotlib.org/stable/tutorials/colors/colormaps.html">ofrece Matplotlib</a>. Supongamos que tenemos la siguiente matriz:

```python
matriz = np.array([[0.1, 0.3], [0.2, 0.1], [0.3, 0.6]])
```

Al llamar a <code>imshow</code> sólo tenemos que pasarle la matriz:

```python
plt.imshow(matriz)
plt.show()
```

Podemos mostrar la escala para que nos hagamos una idea de las magnitudes:

```python
plt.imshow(matriz)
plt.colorbar()
plt.show()
```

Todas las opciones que hemos visto antes para ajustar etiquetas de los ejes, representar en subfiguras, etc. se aplican también a este tipo de gráficas. 

## Integración con Pandas

Todo lo que hemos visto en esta libreta hasta ahora se puede hacer también con objetos <code>Series</code> de Pandas. Adicionalmente, Pandas ofrece muchas de estas funciones ya integradas en los objetos <code>Series</code> y <code>DataFrame</code>. Esto hace que podamos representar rápidamente los datos. Veamos un par de ejemplos:

```python
s = pd.Series(index=np.arange(10), data=np.random.randn(10))
s
```

```python
s.plot()
plt.show()
```

Nuevamente, todos los métodos que hemos visto para personalizar ejes, tamaños, etc. se aplican aquí también; aunque hay algunas limitaciones y algunos casos en los que necesitaremos recurrir a la forma tradicional de llamar a <code>plot()</code> y similares:

```python
plt.plot(s.index, s)
plt.show()
```

Veremos más de la integración con Pandas un poco más adelante. Normalmente, nos convendrá usar esta integración siempre que contenga las funcionalidades que necesitemos. Es especialmente útil cuando simplemente estamos intentando entender mejor los datos que tenemos entre manos.

## Conclusiones

En esta libreta hemos visto una introducción a Matplotlib. Hemos visto los siguientes puntos:
- Representación básica con las funciones <code>plot</code>, <code>bar</code> y <code>scatter</code>.
- Personalización de ejes y representación con subfiguras.
- Representación de matrices con <code>imshow</code>.

Es una de las librerías más complejas con las que nos toparemos, con infinidad de posibilidades, funciones y parámetros de configuración. Por eso, es muy importante tener siempre el <a href="https://matplotlib.org/stable/contents.html">manual</a>, y sobre todo, la <a href="https://matplotlib.org/stable/gallery/index.html">galería</a> a mano.
