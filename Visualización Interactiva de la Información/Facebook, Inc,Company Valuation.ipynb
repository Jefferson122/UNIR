```python
import numpy as np
import matplotlib.pyplot as plt    # Nuevo: paquete Pyplot de Matplotlib
import pandas as pd
from datetime import datetime
```

```python
df=pd.DataFrame()
df['Date']=['12/17/2010','01/10/2011','03/06/2011','06/18/2011','04/13/2011','06/06/2011','06/22/2011','07/06/2011','07/19/2011','07/29/2011','08/16/2011']
df['Valuation ($ Billions)']=['60','66.2','79.2','81.6','76.8','79.2','84','84','84','81.6','79.2']
df['Date'] = pd.to_datetime(df['Date']).dt.date
#df=df.set_index('Date')
df
```

```python
dfnew=pd.DataFrame()
dfnew['Date']=['Dec', 'Jan', 'Feb','March','April','May','June','July','August']
dfnew['Valuation ($ Billions)']=[0,6.2, np.nan ,20.4,16.8,np.nan ,21.6,23.2,19.2]
dfnew
```

```python
x=dfnew['Date']
y=dfnew['Valuation ($ Billions)']
dfnewin=dfnew.interpolate()
eje = dfnewin.plot(marker='.', color='red')
dfnew.plot(ax = eje)
plt.title('Facebook, Inc.:Company Valuation')
plt.xlabel("Date")
plt.ylabel("Valuation ($ Billions)")
plt.plot(x,y,color='blue')
plt.show()
```
