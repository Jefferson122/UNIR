```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def test1(t, y):
    """Función de prueba"""
    return (1+t)/(1+y) # y'=(1+t)/(1+y)
```

```python
def Euler(a, b, y0, f, N):
    """
    Implementación método de Euler
    Entradas:
    a -- inicio intervalo
    b -- fin intervalo
    y0 -- aproximación inicial 
    f -- función
    N -- pasos

    Salida:
    w -- aproximación final
    """
    tvec=list()
    wvec=list()
    h = (b - a)/N
    t = a
    w = y0
    tvec.append(t)
    wvec.append(w)
    print("t0 = {0:.2f}, w0 = {1:.12f}".format(t, w))
    for i in range(1, N+1):
        w = w + h*f(t, w)
        t = a + i*h
        tvec.append(t)
        wvec.append(w)
        print("t{0:<2} = {1:.2f}, w{0:<2} = {2:.12f}".format(i, t, w))
    plt.plot(tvec,wvec)
    return w
```

```python
print("Método de Euler:")
Euler(1, 2, 2, test1, 2) #y0->y(1)=2
```

```python
def test2(t, y):
    """Función de prueba"""
    return -y+t*(y**(1/2)) # y'=-y+t*(y**(1/2))
```

```python
def RK4(a, b, y0, f, N):
    """
    Implementación método RK4
    Entradas:
    a -- inicio intervalo
    b -- fin intervalo
    y0 -- aproximación inicial
    f -- función
    N -- pasos

    Salida:
    w -- aproximación final
    """
    h = (b - a)/N
    t = a
    w = y0
    print("t0 = {0:.2f}, w0 = {1:.12f}".format(t, w))

    for i in range(1, N+1):
        k1 = h*f(t, w)
        k2 = h*f(t + h/2, w + k1/2)
        k3 = h*f(t + h/2, w + k2/2)
        k4 = h*f(t + h, w + k3)
        w = w + (k1 + 2*k2 + 2*k3 + k4)/6
        t = a + i*h
        print("t{0:<2} = {1:.2f}, w{0:<2} = {2:.12f},k1{0:<2} = {2:.12f},k2{0:<2} = {2:.12f},k3{0:<2} = {2:.12f},k4{0:<2} = {2:.12f}".format(i, t, w,k1,k2,k3,k4))
    return w
```

```python
print("Método de Runge kutta 4 orden:")
RK4(2, 3, 2, test2, 5) #y0->y(2)=2
```

```python
def funcion(x):
#Función de prueba
    return np.exp(x)*np.log(x)+(x**2)*(2**x) # retorna pol(x) = x
```

```python
def trapecio(f,a,b,n):
    print('Implementación regla de trapecio')
    """Implementación regla de trapecio
    Entradas:
    f -- función
    a -- inicio intervalo
    b -- fin intervalo
    n -- número de pasos (par)

    Salida:
    raiz -- aproximación área bajo la curva
    """
    h = (b - a)/n
    acum = 0
    print('Valor de f(a)=',f(a),'Valor de f(b)=',f(b),'Valor de h',h)
    for j in range(1, n):
        print('xi=',a+h*j,'donde f(xi)=',f(a + h*j))
        acum += 2*f(a + h*j)
        raiz = (h/2)*(f(a) + acum + f(b))
    print('La suma de todos los elementos f(a+xi) es',acum)
    print('El área es la función es:',raiz)
    return 
```

```python
trapecio(funcion, 2, 3, 10)
```

```python
def simpson(f, a, b, n):
    print('Implementación regla de simpson')
    h = (b - a)/n
    oddsum = 0
    evensum = 0
    print('Valor de f(a)=',f(a),'Valor de f(b)=',f(b),'Valor de h',h)
    for j in range(1, n):
        x = a + h*j
        if j % 2== 0:
            print('par xi=',x,'donde f(xi)=',f(x))
            evensum += f(x) #suma de los valores pares
        else:
            print('impar xi=',x,'donde f(xi)=',f(x))
            oddsum += f(x)  #suma de los valores impares
    raiz = (h/3)*(f(a) + 2*evensum + 4*oddsum + f(b))
    print('Suma de los elementos PARES de f(x)',evensum)
    print('Suma de los elementos IMPARES de f(x)',oddsum)
    print('El área es la función es:',raiz)
```

```python
simpson(funcion, 2, 3, 10)
```
